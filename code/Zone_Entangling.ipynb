```python
import pandas as pd
import numpy as np
```

```python
capacity = pd.read_csv("../data/Monthly_Generating_Capacity_EIA_allSources_LOCALITIES.csv")
capa0416 = pd.read_csv("../data/Annual_Generating_Capacity_EIA_allSources_LOCALITIES.csv")
planned = pd.read_csv("../data/Monthly_Generating_Capacity_EIA_Planned_allSources_LOCALITIES.csv")
```

```python
capacity.columns = [x.replace("Capacity ", "") for x in capacity.columns.to_list()]
capacity['Sum'] = capacity.sum(axis=1)
capacity.set_index('Date',inplace=True)
capacity.head()
```

```python
#create separate df for each zone
filter_col_LI = [col for col in capacity.columns if "LI" in col]
df_LI = capacity[filter_col_LI]
df_LI['Sum (LI)'] = df_LI.iloc[:, -6:-1].sum(axis=1)
filter_col_GJ = [col for col in capacity.columns if "GJ" in col]
df_GJ = capacity[filter_col_GJ]
df_GJ['Sum (GJ)'] = df_GJ.iloc[:, -6:-1].sum(axis=1)
filter_col_NYCA = [col for col in capacity.columns if "NYCA" in col]
df_NYCA = capacity[filter_col_NYCA]
df_NYCA['Sum (NYCA)'] = df_NYCA.iloc[:, -6:-1].sum(axis=1)
filter_col_NYC = [col for col in capacity.columns if "NYC" in col if col not in filter_col_NYCA]
df_NYC = capacity[filter_col_NYC]
df_NYC['Sum (NYC)'] = df_NYC.iloc[:, -6:-1].sum(axis=1)
df_GJ.head()
```

```python
capa0416.columns = [x.replace("Capacity ", "") for x in capa0416.columns.to_list()]
capa0416['Sum'] = capa0416.sum(axis=1)
capa0416.set_index('Date',inplace=True)
capa0416
```

```python
#create separate df for each zone
filter_col_LI = [col for col in capa0416.columns if "LI" in col]
df_capa0416_LI = capa0416[filter_col_LI]
df_capa0416_LI['Sum (LI)'] = df_capa0416_LI.iloc[:, -6:-1].sum(axis=1)
filter_col_GJ = [col for col in capa0416.columns if "GJ" in col]
df_capa0416_GJ = capa0416[filter_col_GJ]
df_capa0416_GJ['Sum (GJ)'] = df_capa0416_GJ.iloc[:, -6:-1].sum(axis=1)
filter_col_NYCA = [col for col in capa0416.columns if "NYCA" in col]
df_capa0416_NYCA = capa0416[filter_col_NYCA]
df_capa0416_NYCA['Sum (NYCA)'] = df_capa0416_NYCA.iloc[:, -6:-1].sum(axis=1)
filter_col_NYC = [col for col in capa0416.columns if "NYC" in col if col not in filter_col_NYCA]
df_capa0416_NYC = capa0416[filter_col_NYC]
df_capa0416_NYC['Sum (NYC)'] = df_capa0416_NYC.iloc[:, -6:-1].sum(axis=1)
df_capa0416_NYCA
```

```python
planned.columns = [x.replace("Capacity ", "") for x in planned.columns.to_list()]
planned['Sum'] = planned.sum(axis=1)
planned.set_index('Date',inplace=True)
planned.head()
```

```python
#create separate df for each zone
filter_col_LI = [col for col in planned.columns if "LI" in col]
df_planned_LI = planned[filter_col_LI]
df_planned_LI['Sum (LI)'] = df_planned_LI.iloc[:, -6:-1].sum(axis=1)
filter_col_GJ = [col for col in planned.columns if "GJ" in col]
df_planned_GJ = planned[filter_col_GJ]
df_planned_GJ['Sum (GJ)'] = df_planned_GJ.iloc[:, -6:-1].sum(axis=1)
filter_col_NYCA = [col for col in planned.columns if "NYCA" in col]
df_planned_NYCA = planned[filter_col_NYCA]
df_planned_NYCA['Sum (NYCA)'] = df_planned_NYCA.iloc[:, -6:-1].sum(axis=1)
filter_col_NYC = [col for col in planned.columns if "NYC" in col if col not in filter_col_NYCA]
df_planned_NYC = planned[filter_col_NYC]
df_planned_NYC['Sum (NYC)'] = df_planned_NYC.iloc[:, -6:-1].sum(axis=1)
df_planned_NYCA.head()
```

## Combine DataFrames to get capacity amounts in real zones

GJ = GJ' + NYC

NYCA = NYCA' + GJ' + NYC + LI

```python
df_NYC_renamed = df_NYC.copy()
df_NYC_renamed.columns = [x.replace("NYC", "GJ") for x in df_NYC_renamed.columns.to_list()]
df_GJ_total = df_GJ.add(df_NYC_renamed,fill_value=0)
df_GJ_total
```

```python
df_GJ_total_renamed = df_GJ_total.copy()
df_GJ_total_renamed.columns = [x.replace("GJ", "NYCA") for x in df_GJ_total_renamed.columns.to_list()]
df_LI_renamed = df_LI.copy()
df_LI_renamed.columns = [x.replace("LI", "NYCA") for x in df_LI_renamed.columns.to_list()]
df_NYCA_total = df_NYCA.add(df_GJ_total_renamed,fill_value=0)
df_NYCA_total = df_NYCA_total.add(df_LI_renamed,fill_value=0)
df_NYCA_total
```

```python
df_capa0416_NYC_renamed = df_capa0416_NYC.copy()
df_capa0416_NYC_renamed.columns = [x.replace("NYC", "GJ") for x in df_capa0416_NYC_renamed.columns.to_list()]
df_capa0416_GJ_total = df_capa0416_GJ.add(df_capa0416_NYC_renamed,fill_value=0)
df_capa0416_GJ_total
```

```python
df_capa0416_GJ_total_renamed = df_capa0416_GJ_total.copy()
df_capa0416_GJ_total_renamed.columns = [x.replace("GJ", "NYCA") for x in df_capa0416_GJ_total_renamed.columns.to_list()]
df_capa0416_LI_renamed = df_capa0416_LI.copy()
df_capa0416_LI_renamed.columns = [x.replace("LI", "NYCA") for x in df_capa0416_LI_renamed.columns.to_list()]
df_capa0416_NYCA_total = df_capa0416_NYCA.add(df_capa0416_GJ_total_renamed,fill_value=0)
df_capa0416_NYCA_total = df_capa0416_NYCA_total.add(df_capa0416_LI_renamed,fill_value=0)
df_capa0416_NYCA_total
```

```python
df_planned_NYC_renamed = df_planned_NYC.copy()
df_planned_NYC_renamed.columns = [x.replace("NYC", "GJ") for x in df_planned_NYC_renamed.columns.to_list()]
df_planned_GJ_total = df_planned_GJ.add(df_planned_NYC_renamed,fill_value=0)
df_planned_GJ_total
```

```python
df_planned_GJ_total_renamed = df_planned_GJ_total.copy()
df_planned_GJ_total_renamed.columns = [x.replace("GJ", "NYCA") for x in df_planned_GJ_total_renamed.columns.to_list()]
df_planned_LI_renamed = df_planned_LI.copy()
df_planned_LI_renamed.columns = [x.replace("LI", "NYCA") for x in df_planned_LI_renamed.columns.to_list()]
df_planned_NYCA_total = df_planned_NYCA.add(df_planned_GJ_total_renamed,fill_value=0)
df_planned_NYCA_total = df_planned_NYCA_total.add(df_planned_LI_renamed,fill_value=0)
df_planned_NYCA_total
```

## Combine monthly and annual operating capacity

```python
df_NYCA_total_all_years = df_capa0416_NYCA_total.append(df_NYCA_total)
df_NYCA_total_all_years
```

```python
df_GJ_total_all_years = df_capa0416_GJ_total.iloc[7:].append(df_GJ_total)
df_GJ_total_all_years
```

```python
df_NYC_all_years = df_capa0416_NYC.iloc[7:].append(df_NYC)
df_NYC_all_years
```

```python
df_LI_all_years = df_capa0416_LI.iloc[7:].append(df_LI)
df_LI_all_years
```

## Save combined DataFrames to csv

```python
df_NYCA_total_all_years.to_csv('../data/Generating_Capacity_NYCA_04-23.csv', index=False)
df_GJ_total_all_years.to_csv('../data/Generating_Capacity_GJ_11-23.csv', index=False)
df_NYC_all_years.to_csv('../data/Generating_Capacity_NYC_11-23.csv', index=False)
df_LI_all_years.to_csv('../data/Generating_Capacity_LI_11-23.csv', index=False)
```
